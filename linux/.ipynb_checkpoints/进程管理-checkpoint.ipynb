- 静态查看系统进程
    - ps命令：
        - 用途： 查看静态的进程统计信息
        - 格式： ps aux
            - 选项
                - a 显示当前终端下的所有进程信息
                - u 使用以用户为主的格式输出进程信息
                - x 显示当前用户在所有终端下的进程信息
            - 显示信息
                - USER  进程属主（用户名）
                - PID   进程唯一编号
                - %CPU  进程占用cpu时间与总时间百分比
                - %MEM  占用总内存与总内存的百分比  
                - VSZ   占用虚拟内存的空间,单位：k
                - RSS   占用总内存空间，单位：k
                - TTY   
                    - 终端 tty1-tty6 本机登录
                    - ？代表与终端无关  
                    - pts/0 代表网络连接
                - STAT  进程状态
                    - R: 代表进程正在运行
                    - S：代表该程序睡眠状态
                    - T：代表该程序已经停止运行或者说侦测
                    - Z: 代表进程已经停止，但依旧使用系统资源（僵尸进程）
                - START 进程启动时间  
                - TIME  进程占用了cpu运作的时间
                - COMMAND 进程有哪个程序发起

- 动态查看系统进程
    - top 
    - 信息解释
        - 第一行
            - top - 
            - 08:49:27  当前系统时间
            - up  2:00  系统启动时间
            - 3 users   当前登陆用户数
            - load average：0.00， 0.01， 0.02 系统的平均负载（每1，5，15的平均负载）
        - 第二行
            - Tasks： 97 total 总进程数
            - 1 running 正在运行的进程数
            - 96 sleeping 睡眠状态运行数
            - 0 stopped stop状态的运行数
            - 0 zombie  僵尸状态的运行数
        - 第三行
            - %Cpu(s)   cpu状态信息  
            - 0.0 us    用户空间占CPU的百分比  
            - 0.0 sy    内核空间占用CPU的百分比  
            - 0.0 ni    改变过优先级的进程占用CPU百分比
            - 100.0 id  空闲CPU百分比  
            - 0.0 wa    等待占用CPU百分比  
            - 0.0 hi    代表硬中断占用CPU的百分比  
            - 0.0 si    软中断占用CPU百分比   
            - 0.0 st
        - 第四行
            - KiB Mem       内存信息
            - 997924 total  总内存
            - 719024 free   空闲内存
            - 135580 used   已经使用内存
            - 143320 buff/cache 缓存的内存量
        - 第五行
            - KiB Swap          交换分区信息   
            - 9433084 total     总量
            - 9433084 free      空闲交换分区量
            - 0 used            已使用的交换分区量
            - 698148 avail Mem  缓冲区交换分区总量
        - 第六行：空行
        - 第七行：
            - PID       进程的ID
            - USER      进程属主（用户名）    
            - PR        进程优先级
            - NI        同上，负值越低优先级越高，负值优先级高于正值优先级
            - VIRT      进程使用的虚拟内存量
            - RES       进程占用的物理内存量
            - SHR       共享内存大小
            - S         进程状态
            - %CPU      CPU时间占用比
            - %MEM      内存再用比
            - TIME+     进程运行时间
            - COMMAND   进程命令
    - 常用选项
        - P 根据CPU占用情况对进程列表进行排序
        - M 根据内存占用情况进程排序
        - N 根据启动时间进程排序
        - h/? 查看帮助信息
        - k 结束进程（9 强制结束进程）
        - r 修改进程优先级（NI）
        - 可以已使用空格强制更新状态信息
        - q 正常退出

- 进程命令
    - kill 终止指定PID号的进程
    - killall 终止所有的进程
    - -9 强制终止
    - pgrep ： 根据特定的条件查找PID信息
        -l 列出进程名称
        -U 根据进程所属的用户进行查找
        -t 根据进程所在的终端进程查找
